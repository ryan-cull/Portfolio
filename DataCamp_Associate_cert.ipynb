## Task 1




```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('food_claims_2212.csv')
```

```python
df.head(5)
df.info
df.dtypes
```

Claim amount should be a float, but other than that, the columns are the right type.
Below, I have converted the data type of the claim amount column to float.
# 

```python
df['claim_amount'] = df['claim_amount'].str.replace('$','')
df['claim_amount'] = df['claim_amount'].str.replace('R','')
df['claim_amount'].head(5)
```

```python
df['claim_amount'] = df['claim_amount'].astype(float)
df.dtypes
```

Now, I will examine the amount of null values within the dataset.

```python
df.isna().sum()
```

The only columns containing any null values are amount paid and linked cases, which contain 36 and 26 null values respectively.

```python
df.dropna(axis=0,inplace= True)
```

Now, I'll double check to make sure all null values were removed.

```python
df.isna().sum()
```

## Task 2

```python
fig = sns.histplot(x=df['location'], color = 'g')
fig.set_xlabel('Location')
plt.show()
```

Recife is the location with the most appearances in our dataset. The observations are not balancec across locations, since Natal and Fortaleza have noticeably less appearances than the other two locations.

## Task 3

```python
fig1 = sns.histplot(x=df['time_to_close'],color='r')
fig1.set_xlabel('Time to Close')
plt.show()
```

The data for time to close follows a relatively normal distrubtion with some outliers toward the right side of the graph. 

## Task 4

```python
sns.boxplot(x=df['location'], y = df['time_to_close'])
```

The median and quartiles for the time to close variable are very similar across locations, but the Sao Luis location seems to contain the most extreme outliers.
